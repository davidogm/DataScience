```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Empezamos con un modelo muy sencillo, red de clasificación binaria.

Este modelo tiene una entrada unidimensional de tamaño 32, la pasa a una capa totalmente conectada con 64 unidades y una función de activación lineal exponencial. Le sigue una última capa densa de una sola neurona con una función de activación sigmoidea.

Tenemos muchas funciones de activación

https://keras.io/api/layers/activations/

image.png

Como siempre la referencia Keras API con toda la documentación

```python
model = Sequential([
    Dense(64, activation='elu', input_shape=(32,)),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
from tensorflow import keras
keras.utils.plot_model(model, show_shapes=False)  # probad con True
```

#### Compilar el método

Optimizadores

Otro sitio estupendo en cuanto a la refencia



https://torres.ai/deep-learning-inteligencia-artificial-keras/


```python
# Define the model optimizer, loss function and metrics
model.compile(
    optimizer='sgd',
    #optimizer=tf.keras.optimizers.SGD(),
    loss='binary_crossentropy',
    metrics=['accuracy', 'mae']
)
```

Definimos otros parámetros para el modelo y vemos la versatilidad que tiene



```python
# Define the model optimizer, loss function and metrics
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# y luego lo de siempre, ver ls resultados con estos atributos

print(model.loss)
print(model.optimizer)
print(model.metrics)
```

También podemos incluir nuevo objeto para optimizar algunos de los parámetros 

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.005)

# e incluyo el optimizador en este punto compilando el modelo
model.compile(
    optimizer=opt,
    loss='categorical_crossentropy',
    metrics=['accuracy', 'mae']
)
```

```python
# e imprimo su optimizador
print (model.optimizer.lr)
```
