```
# =============================================================================
#INTRODUCCIÓN A PANDAS - LECTURA DE ARCHIVOS
# =============================================================================

import pandas as pd
import requests

#EXCEL - CANCER MAMA
#cancer_mama = pd.read_excel("breast_cancer.xlsx")

#CSV (COMMA SEPARATED VALUES) TITANIC Y VINO ROJO
#titanic = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")
#vino_rojo = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",sep=";")

#JSON - COVID19
json_covid = requests.get("https://api.covid19api.com/summary").json()
datos_coronoravirus_dataframe = pd.DataFrame.from_dict(json_covid["Countries"])
#datos_coronoravirus = pd.read_csv("covid_20200430.csv")

datos_coronoravirus_dataframe.info()

#vino_rojo.describe()
#sample_cancer = cancer_mama.head(10)
#titanic_sample = titanic.tail(50)
```

```
import matplotlib.pyplot as plt

labels = Sub.CountryCode        # valores eje x
Confirmed = Sub.TotalConfirmed  # valores eje y


#fig, ax = plt.subplots(figsize=(14,6), dpi=300)

fig, ax = plt.subplots(figsize=(14,6))

ax.bar(labels, Confirmed, width=1, label='Confirmed')

ax.set_ylabel('Confirmed')
ax.set_title('Countries')
ax.legend()

ax.grid()

plt.show()

```

```
import matplotlib.pyplot as plt

valoresX = ("Alto","Bajo")  
#print(valoresX)

valoresY=pd.value_counts(vino_rojo['alcohol']>12)
plt.bar(valoresX,valoresY)
#plt.xticks(months)
#plt.ylabel('Sales in USD ($)')
#plt.xlabel('Month number')
plt.show()
```

```
vino_rojo[["alcohol","quality"]]    #comprobar la diferencia y lo bien que se ve con el doble corchete

vr=vino_rojo.sort_values(by=['alcohol'],ascending=False)
vr.head(10)

vr_sub=vr.head(20)
```

```
vr_sub.index
```

```
import matplotlib.pyplot as plt

# con esto no se ve bien porque son demasiados vinos
#vino = vr.index
#quality = vr.quality

# con el subconjunto se ve más claro
vino = vr_sub.index
quality = vr_sub.quality


fig, ax = plt.subplots(figsize=(30,5))        # con 30 coge el máximo

ax.bar(vino, quality, width=1, label='Vinos')

ax.set_ylabel('quality')
ax.set_title('vino')
ax.legend()


plt.show()
```
